```python
orders_delivered = orders[orders['order_status'] == 'delivered']
```

```python
print(payments.isnull().sum())
print(payments.duplicated().sum())
```

```python
payment_summary = payments.groupby("order_id")["payment_value"].sum().reset_index()
```

```python
payment_summary = payment_summary.rename(columns={"payment_value": "total_payment"})
```

```python
print(type(payment_summary))
print(type(orders_delivered))
```

```python
orders_enriched = orders_delivered.merge(
    payment_summary,
    on="order_id",
    how="left"
)
```

```python
payment_summary
```

```python
orders_enriched
```

```python
orders_enriched['order_purchase_timestamp']= pd.to_datetime(orders_enriched['order_purchase_timestamp'])
```

```python
orders_enriched['month']= orders_enriched['order_purchase_timestamp'].dt.to_period('M')
orders_enriched['date']= orders_enriched['order_purchase_timestamp'].dt.date
```

```python
orders_enriched
```

```python
orders_enriched = orders_enriched.merge(
    customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)
```

```python
customer_metrics = orders_enriched.groupby("customer_unique_id").agg({
    "order_id": "count",
    "total_payment": "sum"
}).reset_index()
customer_metrics = customer_metrics.rename(columns={
    "order_id": "order_count",
    "total_payment": "total_spent"
})
```

```python
latest_date = orders_enriched['order_purchase_timestamp'].max()
```

```python
recency = orders_enriched.groupby("customer_unique_id")[
    "order_purchase_timestamp"
].max().reset_index(
    )   
recency["recency_days"] = (
    latest_date - recency["order_purchase_timestamp"]
).dt.days
```

```python
customer_metrics = customer_metrics.merge(
    recency[['customer_unique_id', 'recency_days']],
    on='customer_unique_id'
)
```

```python
customer_metrics
```

```python
customer_metrics['customer_type'] = customer_metrics['order_count'].apply(
    lambda x: 'repeat' if x > 1 else 'one-time'
)
```

```python
customer_metrics.head()
customer_metrics['order_count'].value_counts()
```

```python
customer_metrics.to_csv('customer_metrics.csv', index=False)
orders_enriched.to_csv('orders_enriched.csv', index=False)
```

```python
orders_enriched['date'] = pd.to_datetime(orders_enriched['date'])

orders_enriched['year'] = orders_enriched['date'].dt.year
orders_enriched['month'] = orders_enriched['date'].dt.month
```

```python
base_path = "partitioned_data/"

for year in orders_enriched['year'].unique():
    for month in orders_enriched[orders_enriched['year'] == year]['month'].unique():
        subset = orders_enriched[(orders_enriched['year'] == year) & (orders_enriched['month'] == month)]
        
        # 🔥 REMOVE partition columns
        subset = subset.drop(columns=['year', 'month'])
        
        path = f"{base_path}year={year}/month={month}/"
        os.makedirs(path, exist_ok=True)
        
        subset.to_parquet(f"{path}/data.parquet", index=False)
```
